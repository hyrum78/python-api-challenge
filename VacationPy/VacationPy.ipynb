```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Import API key
from api_keys import g_key
```

### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
weather_import = pd.read_csv('../WeatherPy/output_data/cities.csv')
weather_data = weather_import.set_index('City ID')
weather_data.head()
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Store latitude and longitude in locations
lat_long = weather_data[['Lat', 'Lng']]

# Fill NaN values and convert to float
humid = weather_data["Humidity"]
lat_long.head()
```

```python
# plot heat map
fig = gmaps.figure(center = (0,0),zoom_level=2)

heat_map = gmaps.heatmap_layer(lat_long, weights = humid, dissipating=False, max_intensity = 100, point_radius = 1.5)

fig.add_layer(heat_map)
fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
# where is it nice?
nice_temps = (weather_data['Max Temp'] < 80) & (weather_data['Max Temp'] > 70)
low_windspeeds = weather_data['Wind Speed'] < 10
clear_skies = weather_data['Cloudiness'] == 0
lets_get_away = nice_temps & low_windspeeds & clear_skies
```

```python
# Use indexing to filter weather_data
lets_get_away_df = weather_data[lets_get_away]
lets_get_away_df = lets_get_away_df.dropna()
lets_get_away_df
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
hotel_df = lets_get_away_df.loc[:,['City','Country','Lat','Lng']]

# Add column Hotel Name
hotel_df['Hotel Name'] = ""
hotel_df
```

```python
# params dictionary
params = {
    "radius": 5000,
    "type": "hotel",
    "keyword": "hotel",
    "key": g_key
}

# lat/lng to find hotels
for index, row in hotel_df.iterrows():
    lat = row["Lat"]
    lng = row["Lng"]

    params["location"] = F"{lat},{lng}"
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

    nearby_search = requests.get(base_url, params=params)

    # json
    nearby_search = nearby_search.json()

    # try/except to skip missing data points
    try:
        hotel_df.loc[index, "Hotel Name"] = nearby_search["results"][0]["name"]
        
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
#hotel_df
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lng"]]
```

```python
# Add marker layer ontop of heat map

add_hotels = gmaps.marker_layer(locations, info_box_content = hotel_info)

fig.add_layer(add_hotels)

fig
```
